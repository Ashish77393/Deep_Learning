```python
import numpy as np
import pandas as pd 
```

```python
df=pd.DataFrame([[8,8,4],[7,9,7],[5,12,17]],columns=['cgpa','profile_score','salary(lpa)'])
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(2,activation='linear',input_dim=2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['salary(lpa)'].values,epochs=500,verbose=1,batch_size=1)
```
