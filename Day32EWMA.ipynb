```python
import pandas as pd 
```

```python
df=pd.read_csv("DailyDelhiClimateTest.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['date'],df['meantemp'],color='red')
plt.show()
```

```python
x1=df['meantemp'].ewm(alpha=1).mean()
x1
```

```python
df['ewm']=x1
df
```

```python
plt.scatter(df['date'],df['meantemp'],color="red")
plt.plot(df['date'],x1,color='black',label='alpha=0.9')
plt.legend()
plt.show()
```
